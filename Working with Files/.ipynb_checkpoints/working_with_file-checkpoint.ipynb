# Working with Files in Python
## 1. What is File Handling?

File handling allows us to:

* Store data permanently (in files)

* Read data from files

* Write or modify data in files

### Python provides built-in functions to handle files easily.

## 2. Types of Files

- 1. Text Files (.txt, .csv, .py)

- Store data as plain text

- 2. Binary Files (.jpg, .mp3, .dat)

- Store data in binary format (0s and 1s)

## 3. Opening a File

We use the `open()` function:

`file = open("example.txt", "mode")`

Modes of Opening File
| Mode | Meaning           |
| ---- | ----------------- |
| `r`  | Read (default)    |
| `w`  | Write (overwrite) |
| `a`  | Append            |
| `x`  | Create new file   |
| `b`  | Binary mode       |
| `t`  | Text mode         |


## 4. Reading from File
**Example 1: Read entire file**

```python
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()
```

**Example 2: Read line by line**

```python
file = open("data.txt", "r")

for line in file:
    print(line)

file.close()
```

**Example 3: Read specific lines**

```python
file = open("data.txt", "r")

print(file.readline())  # first line
print(file.readline())  # second line

file.close()
```

## 5. Writing to File
**Example 1: Write (overwrite)**

```python
file = open("data.txt", "w")
file.write("Hello Students!\n")
file.write("Welcome to Python File Handling.")
file.close()
```

Old data will be deleted

**Example 2: Append data**

```python
file = open("data.txt", "a")
file.write("\nThis line is added later.")
file.close()
```

## 6. Using with Statement (Best Practice)

**Automatically closes the file**

```python
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

No need to use `close()`

## 7. Writing Multiple Lines

```python
lines = ["Python\n", "Java\n", "C++\n"]

with open("data.txt", "w") as file:
    file.writelines(lines)

# Read and print the contents
with open("data.txt", "r") as file:
    contents = file.read()
    print(contents)
```

## 8. File Pointer Methods

| Function  | Description               |
| --------- | ------------------------- |
| `tell()`  | Returns current position  |
| `seek(n)` | Moves pointer to position |

**Example**

```python
with open("data.txt", "r") as file:
    print(file.tell())   # position
    file.read(6)
    # print(file.read())
    print(file.tell())   # updated position
    
    file.seek(0)         # go back to start
    print(file.read())
```

## 9. Checking if File Exists

```python
import os

if os.path.exists("data.txt"):
    print("File exists")
else:
    print("File not found")
```

## 10. Deleting a File

```python
import os

os.remove("data.txt")
```

## 11. Practical Example (Student Record System)

```python
# Writing student data
with open("students.txt", "w") as file:
    file.write("1, Ayush, 85\n")
    file.write("2, Rahul, 90\n")

# Reading student data
with open("students.txt", "r") as file:
    for line in file:
        data = line.split(",")
        print("ID:", data[0])
        print("Name:", data[1])
        print("Marks:", data[2])
        print("------")
```

## 12. Important Points for

✔ Always close files (or use with)

✔ Use correct mode (r, w, a)

✔ Handle errors (file not found, etc.)

✔ Prefer with statement (safe and clean)

## 13. Common Errors

❌ FileNotFoundError

❌ PermissionError

**Example Handling Error:**

```python
try:
    with open("data.txt", "r") as file:
        print(file.read())
except FileNotFoundError:
    print("File does not exist!")
```
