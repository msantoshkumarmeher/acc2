# Working with Files in Python
## 1. What is File Handling?

File handling allows us to:

* Store data permanently (in files)

* Read data from files

* Write or modify data in files

### Python provides built-in functions to handle files easily.

## 2. Types of Files

1. Text Files (.txt, .csv, .py)

    - Store data as plain text

2. Binary Files (.jpg, .mp3, .dat)

    - Store data in binary format (0s and 1s)

## 3. Opening a File

We use the `open()` function:

`file = open("example.txt", "mode")`

Modes of Opening File
| Mode | Meaning           |
| ---- | ----------------- |
| `r`  | Read (default)    |
| `w`  | Write (overwrite) |
| `a`  | Append            |
| `x`  | Create new file   |
| `b`  | Binary mode       |
| `t`  | Text mode         |


## 4. Reading from File
**Example 1: Read entire file**

```python
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()
```

**Example 2: Read line by line**

```python
file = open("data.txt", "r")

for line in file:
    print(line)

file.close()
```

**Example 3: Read specific lines**

```python
file = open("data.txt", "r")

print(file.readline())  # first line
print(file.readline())  # second line

file.close()
```

## 5. Writing to File

## 6. Using with Statement (Best Practice)

**Automatically closes the file**

```python
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
```

No need to use `close()`

## 7. Writing Multiple Lines

```python
lines = ["Python\n", "Java\n", "C++\n"]

with open("data.txt", "w") as file:
    file.writelines(lines)

# Read and print the contents
with open("data.txt", "r") as file:
    contents = file.read()
    print(contents)
```

## 8. Checking if File Exists

```python
import os

if os.path.exists("data.txt"):
    print("File exists")
else:
    print("File not found")
```

## 9. Deleting a File

```python
import os

os.remove("data.txt")
```

## 10. Practical Example (Student Record System)

```python
# Writing student data
with open("students.txt", "w") as file:
    file.write("1, Ayush, 85\n")
    file.write("2, Rahul, 90\n")

# Reading student data
with open("students.txt", "r") as file:
    for line in file:
        data = line.split(",")
        print("ID:", data[0])
        print("Name:", data[1])
        print("Marks:", data[2])
        print("------")
```

## 11. Important Points for

✔ Always close files (or use with)

✔ Use correct mode (r, w, a)

✔ Handle errors (file not found, etc.)

✔ Prefer with statement (safe and clean)

## 13. Common Errors

# Advanced File Handling in Python

## 1. Working with CSV Files

CSV (Comma Separated Values) is widely used for storing structured data.

Python provides a built-in module: `csv` module

**Example: Writing CSV File**

```python
import csv

with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    
    writer.writerow(["ID", "Name", "Marks"])
    writer.writerow([1, "Santosh", 85])
    writer.writerow([2, "Rahul", 90])
```

## Example: Reading CSV File

```python
import csv

with open("students.csv", "r") as file:
    reader = csv.reader(file)
    
    for row in reader:
        print(row)
```

## Using Dictionary in CSV

```python
import csv

data = [
    {"id": 1, "name": "Santosh", "marks": 85},
    {"id": 2, "name": "Rahul", "marks": 90}
]

with open("students.csv", "w", newline="") as file:
    writer = csv.DictWriter(file, fieldnames=["id", "name", "marks"])
    
    writer.writeheader()
    writer.writerows(data)
```

## 2. Working with JSON Files

JSON is used in APIs and web applications.

Python module: `json` module

**Example: Writing JSON**

```python
import json

data = {
    "name": "Santosh",
    "marks": 85
}

with open("data.json", "w") as file:
    json.dump(data, file)
```

**Example: Reading JSON**

```python
import json

with open("data.json", "r") as file:
    data = json.load(file)
    print(data["name"], data["marks"])
```

## 3. Binary File Handling

Used for images, videos, etc.

**Example: Writing Binary File**

```python
with open("file.bin", "wb") as file:
    file.write(b"Hello World")
```

**Example: Reading Binary File**

```python
with open("file.bin", "rb") as file:
    data = file.read()
    print(data)
```

## 4. Exception Handling in Files

Helps prevent program crashes.

**Example:**

```python
try:
    with open("data.txt", "r") as file:
        print(file.read())
        
except FileNotFoundError:
    print("File not found")

except PermissionError:
    print("No permission to access file")

finally:
    print("Execution completed")
```

## 5. File Handling with User Input

**Example**

```python
filename = input("Enter file name: ")

try:
    with open(filename, "r") as file:
        print(file.read())
        
except FileNotFoundError:
    print("File does not exist!")
```

## 6. Mini Project: Student File System
Features:
* Add student
* View students

**Code :**

```python
def add_student():
    with open("students.txt", "a") as file:
        name = input("Enter name: ")
        marks = input("Enter marks: ")
        file.write(name + "," + marks + "\n")

def view_students():
    with open("students.txt", "r") as file:
        for line in file:
            name, marks = line.strip().split(",")
            print("Name:", name, "Marks:", marks)

while True:
    print("\n1. Add Student")
    print("2. View Students")
    print("3. Exit")
    
    choice = input("Enter choice: ")
    
    if choice == "1":
        add_student()
    elif choice == "2":
        view_students()
    elif choice == "3":
        break
    else:
        print("Invalid choice")
```
