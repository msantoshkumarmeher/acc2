## 1. Password Generator

**Automatically generate strong passwords**

```python
import random
import string

length = 8
characters = string.ascii_letters + string.digits + "!@#$"

password = "".join(random.choice(characters) for i in range(length))
print("Generated Password:", password)
```

## 2. Bulk File Organizer

**Automatically organize files into folders**

```python
import os
import shutil

files = os.listdir()

for file in files:
    if file.endswith(".gif"):
        os.makedirs("Gifs", exist_ok=True)
        shutil.move(file, "Gifs")
    elif file.endswith(".txt"):
        os.makedirs("Text", exist_ok=True)
        shutil.move(file, "Text")
```

## 3. Auto Email Sender (Basic)

```python
import smtplib

server = smtplib.SMTP("smtp.gmail.com", 587)
server.starttls()

server.login("< Write Your Gmail Here>", "<Your Gmail App Password>")

message = "Hello, this is email automation taught in class!"
# Method 2: More compact using list comprehension
# emails = [f"{roll}cse@gcekjr.ac.in" for roll in range(2401104061, 2401104139) if roll != 2401104090]
emails = ["recepent 1 email","recepent 2 email", "recepent 3 email"]
for email in emails:
    server.sendmail("< Write Your Gmail Here>", email, message)

server.quit()
```

## 5. Automate Keyboard

```python
import pyautogui
import time

time.sleep(5)
for i in range (5):
    pyautogui.write("Hello Students!", interval=0.2)
    pyautogui.press("enter")
```

## 6.Build Your Own “Jarvis”

```python

import pyttsx3
import webbrowser

engine = pyttsx3.init()

command = input("Enter command: ")

if "open google" in command:
    webbrowser.open("https://google.com")
    engine.say("Opening Google")

engine.runAndWait()
```
