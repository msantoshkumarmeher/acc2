# Classes and Objects in Python

## 1. What is a Class?

A class is a blueprint or template used to create objects.

**Think of it like:**

Class = Design of a house

Object = Actual house built using that design

**Syntax:**

```python
class ClassName:
    # attributes (variables)
    # methods (functions)
    pass
```

## 2. What is an Object?

An object is an instance of a class.

It represents a real-world entity.

**Example:**

```python
class Student:
    pass

s1 = Student()   # object created
```

## 3. Class with Attributes and Methods
**Example:**

```python
class Student:
    def display(self):
        print("Hello Student")

s1 = Student()
s1.display()
```

## 4. Understanding self

`self` refers to the current object

- It is used to access variables and methods inside the class
- It must be the first parameter in methods

## 5. Adding Data (Attributes)
**Example:**

```python
class Student:
    def set_data(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student()
s1.set_data("Santosh", 85)
s1.display()
```

## 6. Constructor (`__init__` method)

A constructor is a special method that runs automatically when an object is created.

**Example:**

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

s1 = Student("Santosh", 90)
s1.display()
```

## 7. Multiple Objects

```python
s1 = Student("Ram", 80)
s2 = Student("Shyam", 85)

s1.display()
s2.display()
```

## 8. Real-Life Example (Very Important for Students)

```python
class Car:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def show(self):
        print("Brand:", self.brand)
        print("Speed:", self.speed)

c1 = Car("BMW", 200)
c2 = Car("Audi", 180)

c1.show()
c2.show()
```

## Example 1: Student Management System

This example has multiple functions:

- set data
- display
- calculate grade
- check pass/fail

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print("Name:", self.name)
        print("Marks:", self.marks)

    def calculate_grade(self):
        if self.marks >= 90:
            print("Grade: O")
        elif self.marks >= 75:
            print("Grade: A")
        elif self.marks >= 50:
            print("Grade: B")
        else:
            print("Grade: Fail")

    def is_pass(self):
        if self.marks >= 50:
            print("Result: Pass")
        else:
            print("Result: Fail")


# Creating object
s1 = Student("Santosh", 82)

s1.display()
s1.calculate_grade()
s1.is_pass()
```

## Example 2: Bank Account System

**Functions:**
- deposit
- withdraw
- check balance

```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print("Deposited:", amount)

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient balance")
        else:
            self.balance -= amount
            print("Withdrawn:", amount)

    def check_balance(self):
        print("Account Holder:", self.name)
        print("Current Balance:", self.balance)


# Object
acc1 = BankAccount("Santosh", 5000)

acc1.deposit(2000)
acc1.withdraw(1000)
acc1.check_balance()
```

## Example 3: Calculator Class

Multiple operations in one class

```python
class Calculator:
    def add(self, a, b):
        print("Addition:", a + b)

    def subtract(self, a, b):
        print("Subtraction:", a - b)

    def multiply(self, a, b):
        print("Multiplication:", a * b)

    def divide(self, a, b):
        if b == 0:
            print("Cannot divide by zero")
        else:
            print("Division:", a / b)


# Object
c = Calculator()

c.add(10, 5)
c.subtract(10, 5)
c.multiply(10, 5)
c.divide(10, 5)
```

## Example 4: Employee System

**Functions:**

- Display
- Annual salary
- Bonus calculation

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display(self):
        print("Name:", self.name)
        print("Monthly Salary:", self.salary)

    def annual_salary(self):
        print("Annual Salary:", self.salary * 12)

    def bonus(self):
        bonus = self.salary * 0.10
        print("Bonus:", bonus)
        


# Object
e1 = Employee("Rahul", 30000)

e1.display()
e1.annual_salary()
e1.bonus()
```

# Inheritance in Python

## 1. What is Inheritance?

Inheritance allows one class to use properties and methods of another class

- Parent Class (Base class)
- Child Class (Derived class)

Real-life example:

- Parent: Vehicle
- Child: Car, Bike

### Basic Example

```python
class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):   # Inheriting Animal class
    def bark(self):
        print("Dog barks")


d = Dog()
d.speak()   # inherited method
d.bark()
```

### Types of Inheritance

- Single Inheritance
- Multiple Inheritance
- Multilevel Inheritance

**1. Single Inheritance**

One child class inherits from one parent class

```python
class Parent:
    def show(self):
        print("This is Parent class")

class Child(Parent):
    def display(self):
        print("This is Child class")


obj = Child()
obj.show()      # inherited
obj.display()
```

**2. Multiple Inheritance**

One child class inherits from multiple parent classes

```python
class Father:
    def skills(self):
        print("Father: Gardening")

class Mother:
    def talent(self):
        print("Mother: Cooking")

class Child(Father, Mother):
    def show(self):
        print("Child has both skills")


obj = Child()
obj.skills()
obj.talent()
obj.show()
```

**3. Multilevel Inheritance**

Chain of inheritance (Grandparent → Parent → Child)

```python
class Grandparent:
    def house(self):
        print("Grandparent's House")

class Parent(Grandparent):
    def car(self):
        print("Parent's Car")

class Child(Parent):
    def bike(self):
        print("Child's Bike")


obj = Child()
obj.house()
obj.car()
obj.bike()
```

## 2. Method Overriding

When a child class changes the method of parent class

**Example**

```python
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):   # overriding
        print("Dog barks")


d = Dog()
d.speak()
```

## 3. Encapsulation
What is Encapsulation?
- Wrapping data and methods together
- Restricting direct access to variables

**Types of Variables**
* Public → name
* Protected → _name
* Private → __name
  
**Example**

```python
class Student:
    def __init__(self, name, age, marks):
        # Public variable
        self.name = name
        
        # Protected variable
        self._age = age
        
        # Private variable
        self.__marks = marks

    # Public method
    def display(self):
        print("Name (Public):", self.name)
        print("Age (Protected):", self._age)
        print("Marks (Private):", self.__marks)

    # Setter for private variable
    def set_marks(self, m):
        self.__marks = m

    # Getter for private variable
    def get_marks(self):
        return self.__marks


# Creating object
s = Student("Santosh", 22, 80)

# Accessing public variable
print("Public:", s.name)

# Accessing protected variable (possible, but not recommended)
print("Protected:", s._age)

# Accessing private variable (will give error if uncommented)
# print(s.__marks)

# Access private variable using getter
print("Private (using getter):", s.get_marks())

# Modifying private variable using setter
s.set_marks(90)
print("Updated Marks:", s.get_marks())

# Display all
s.display()
```

## 4. Polymorphism

**Same function name behaves differently**

```python
class Bird:
    def sound(self):
        print("Bird makes sound")

class Sparrow(Bird):
    def sound(self):
        print("Sparrow chirps")

class Crow(Bird):
    def sound(self):
        print("Crow caws")


for b in (Sparrow(), Crow(),Bird()):
    b.sound()
```

## 5. Super Keyword `super()`

Used to call parent class constructor or methods

```python
class Person:
    def __init__(self, name):
        self.name = name

class Student(Person):
    def __init__(self, name, marks):
        # Person.__init__(self, name)
        super().__init__(name)
        self.marks = marks

    def display(self):
        print(self.name, self.marks)


s = Student("Santosh", 90)
s.display()
```

## Direct Parent Call vs super()
Let’s take multiple inheritance

Using `Parent.__init__()` **(Problem)**

```python
class A:
    def __init__(self):
        print("A constructor")

class B(A):
    def __init__(self):
        A.__init__(self)
        print("B constructor")

class C(A):
    def __init__(self):
        A.__init__(self)
        print("C constructor")

class D(B, C):
    def __init__(self):
        B.__init__(self)
        C.__init__(self)
        print("D constructor")

d = D()
```

### ❌ Problem:
- A constructor is called twice
- This causes:
  * Duplicate work
  * Bugs in real applications (like DB connections, file loading, etc.)

## Solution Using `super()`

```python
class A:
    def __init__(self):
        print("A constructor")

class B(A):
    def __init__(self):
        super().__init__()
        print("B constructor")

class C(A):
    def __init__(self):
        super().__init__()
        print("C constructor")
        

class D(B, C):
    def __init__(self):
        super().__init__()
        print("D constructor")

# print(D.mro())  # Method Resolution Order
d = D()
```

# Solutions

**1. Single Inheritance**

```python
class Person:
    def show_name(self):
        print("Name: Santosh")

class Student(Person):
    def show_marks(self):
        print("Marks: 85")

s = Student()
s.show_name()
s.show_marks()
```

**2. Multiple Inheritance**

```python
class Teacher:
    def teach(self):
        print("Teaching students")

class Researcher:
    def research(self):
        print("Doing research")

class Professor(Teacher, Researcher):
    def role(self):
        print("Professor role")

p = Professor()
p.teach()
p.research()
p.role()
```

**3. Multilevel Inheritance**

```python
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    def drive(self):
        print("Car is driving")

class ElectricCar(Car):
    def charge(self):
        print("Charging battery")

e = ElectricCar()
e.start()
e.drive()
e.charge()
```

**4. Method Overriding**

```python
class Animal:
    def sound(self):
        print("Animal makes sound")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

c = Cat()
c.sound()
```

**5. Encapsulation**

```python
class Bank:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("Insufficient balance")

    def show_balance(self):
        print("Balance:", self.__balance)


b = Bank()
b.deposit(1000)
b.withdraw(500)
b.show_balance()
```

**6. Polymorphism**

```python
class Shape:
    def area(self):
        print("Calculate area")

class Circle(Shape):
    def area(self):
        print("Circle Area:", 3.14 * 5 * 5)

class Square(Shape):
    def area(self):
        print("Square Area:", 4 * 4)

for s in (Circle(), Square()):
    s.area()
```
