# BeautifulSoup

BeautifulSoup is a Python library used for web scraping. It helps extract data from HTML and XML documents easily by parsing the webpage structure.

## What is Web Scraping?

Web scraping means automatically extracting data from websites.

**Example tasks:**

- Collect product prices from e-commerce sites

- Extract news headlines

- Get table data from websites

- Collect links or images

- BeautifulSoup helps you navigate HTML elements like tags, classes, ids, attributes etc.

## First BeautifulSoup Program
**Example: Parsing simple HTML**

```python
from bs4 import BeautifulSoup

html_doc = """
<html>
<head><title>My Website</title></head>
<body>
<h1>Hello Students</h1>
<p>This is a python class</p>
</body>
</html>
"""

soup = BeautifulSoup(html_doc, "html.parser")

print(soup.title)
print(soup.h1)
print(soup.p)
```

## Extract Only Text

**To get only text inside tags:**

```python
print(soup.title.text)
print(soup.h1.text)
print(soup.p.text)
```

## Finding Elements (find)

**`find()`** returns the first matching tag.

```python
from bs4 import BeautifulSoup

html = """
<html>
<body>
<h1>Python</h1>
<h1>Flask</h1>
<h1>Django</h1>
</body>
</html>
"""

soup = BeautifulSoup(html,"html.parser")

result = soup.find("h1")

print(result.text)
```

## Finding Multiple Elements (find_all)

**`find_all()`** returns all matching tags.

```python
from bs4 import BeautifulSoup

html = """
<html>
<body>
<h1>Python</h1>
<h1>Flask</h1>
<h1>Django</h1>
</body>
</html>
"""

soup = BeautifulSoup(html,"html.parser")

data = soup.find_all("h1")

for item in data:
    print(item.text)
```

## Finding Elements by Class

```python
from bs4 import BeautifulSoup

html = """
<p class="course">Python</p>
<p class="course">Flask</p>
<p class="course">Django</p>
"""

soup = BeautifulSoup(html,"html.parser")

courses = soup.find_all("p", class_="course")

for c in courses:
    print(c.text)
```

## Finding Elements by ID

```python
html = '<h1 id="title">Welcome</h1>'

soup = BeautifulSoup(html,"html.parser")

data = soup.find(id="title")

print(data.text)
```

## Extracting Links from HTML

```python
from bs4 import BeautifulSoup

html = """
<a href="https://google.com">Google</a>
<a href="https://youtube.com">YouTube</a>
"""

soup = BeautifulSoup(html,"html.parser")

links = soup.find_all("a")

for link in links:
    print(link.get("href"))
```

## Scraping Real Website

**Example:** Get title of a webpage.

```python
import requests
from bs4 import BeautifulSoup

url = "http://gcekjr.ac.in"

response = requests.get(url)

soup = BeautifulSoup(response.text,"html.parser")

print(soup.title.text)
```

**Explanation:**

| Code            | Meaning           |
| --------------- | ----------------- |
| requests.get()  | downloads webpage |
| response.text   | HTML source       |
| BeautifulSoup() | parse HTML        |
| soup.title      | access title      |


## Extract All Headlines from Website

**Example:**

```python
import requests
from bs4 import BeautifulSoup

url = "http://gcekjr.ac.in"

response = requests.get(url)

soup = BeautifulSoup(response.text,"html.parser")

headlines = soup.find_all("h2")

for h in headlines:
    print(h.text)
```

## Extract Table Data

```python
from bs4 import BeautifulSoup

html = """
<table>
<tr><td>Python</td><td>100</td></tr>
<tr><td>Flask</td><td>80</td></tr>
</table>
"""

soup = BeautifulSoup(html,"html.parser")

rows = soup.find_all("tr")

for row in rows:
    data = row.find_all("td")

    for item in data:
        print(item.text)
```

## Using CSS Selectors

BeautifulSoup supports CSS selectors.

```python
# Install libraries (run once if not installed)
# !pip install beautifulsoup4
# !pip install requests

from bs4 import BeautifulSoup
html = """
<html>
<head>
<title>Student Courses</title>
</head>

<body>

<h1 id="main-title">Programming Courses</h1>

<div class="course">
<p class="name">Python</p>
<p class="price">5000</p>
</div>

<div class="course">
<p class="name">Flask</p>
<p class="price">4000</p>
</div>

<div class="course">
<p class="name">Django</p>
<p class="price">6000</p>
</div>

<a href="https://python.org">Python Official</a>
<a href="https://flask.palletsprojects.com">Flask Website</a>

</body>
</html>
"""

soup = BeautifulSoup(html, "html.parser")
```

## Select Tag

**Select all `<p>` tags.**

```python
data = soup.select("p")

for item in data:
    print(item.text)
```

## Select Class

Use `.` before class name.

```python
courses = soup.select(".name")

for course in courses:
    print(course.text)
```

## Select ID

Use `#` before id.

```python
title = soup.select("#main-title")

for t in title:
    print(t.text)
```

## Select Inside Tag

Select elements inside another element.

```python
course_names = soup.select("div.course p.name")

for name in course_names:
    print(name.text)
```

## Select Links

```python
links = soup.select("a")

for link in links:
    print(link.get("href"))
```

## Select First Element

```python
first_course = soup.select_one(".name")

print(first_course.text)
```
