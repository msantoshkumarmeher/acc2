## 1. Introduction to NumPy

NumPy (Numerical Python) is a Python library used for scientific computing and numerical operations. It provides support for:
- Large multi-dimensional arrays
- Mathematical operations
- Linear algebra
- Random number generation
- Statistical operations
The main object of NumPy is ndarray (N-dimensional array).

### Why NumPy is faster than Python lists?
- Uses C language implementation internally
- Stores data in continuous memory
- Performs vectorized operations

**Example :**

```python
import numpy as np

a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

print(a + b)
```

In normal Python lists this would require loops.

## 3. Creating NumPy Arrays
### 3.1 Creating Array from Python List

```python
import numpy as np
arr = np.array([1,2,3,4,5])
print(arr)
```

### 3.2 Creating 2D Array

```python
import numpy as np
arr = np.array([[1,2,3],
                [4,5,6]])
print(arr)
```

## 4. Dimensions of Array

| Dimension | Example           |
| --------- | ----------------- |
| 1D        | `[1,2,3]`         |
| 2D        | `[[1,2],[3,4]]`   |
| 3D        | `[[[1,2],[3,4]]]` |


```python
arr = np.array([1,2,3])
print(arr.ndim)
```

## 5. Array Attributes

Important attributes of NumPy array:

| Attribute | Description          | 
| --------- | -------------------- |
| ndim      | number of dimensions |
| shape     | size of array        |
| size      | total elements       |
| dtype     | datatype             |
| itemsize  | memory size          |


```python
# Example:  
import numpy as np
arr = np.array([[1,2,3],[4,5,6]])
print(arr.ndim)
print(arr.shape)
print(arr.size)
print(arr.dtype)
print(arr.itemsize)
```

## 6. Creating Special Arrays
### 6.1 Zeros Array

Creates array filled with zeros.

```python
arr = np.zeros((2,3))
print(arr)
```

### 6.2 Ones Array

```python
arr = np.ones((3,3))
print(arr)
```

### 6.3 Identity Matrix

```python
arr = np.eye(3)
print(arr)
```

### 6.4 Arange Function
Works like range().

```python
arr = np.arange(1,10)
print(arr)
```

### 6.5 Linspace
Creates equally spaced numbers.

```python
arr = np.linspace(0,10,5)
print(arr)
```

## 7. Array Indexing

Access elements using index.

**Example:**

```python
arr = np.array([10,20,30,40])

print(arr[0])
print(arr[2])
```

**2D Indexing**

```python
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr[0,1])
```

## 8. Array Slicing

Used to extract part of array.

Syntax : `array[start:end]`

```python
arr = np.array([1,2,3,4,5,6])

print(arr[1:4])
```

**2D Slicing**

```python
arr = np.array([[1,2,3],
                [4,5,6]])

print(arr[:,1])
```

## 9. Reshaping Arrays

Changes the shape of array.

**Example:**

```python
arr = np.array([1,2,3,4,5,6])

new_arr = arr.reshape(2,3)

print(new_arr)
```

## 10. Mathematical Operations

NumPy performs element-wise operations.

**Example**

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a - b)
print(a * b)
print(a / b)
```

## 11. Aggregate Functions

Used to perform statistical operations.

| Function  | Purpose            |
| --------- | ------------------ |
| np.sum()  | sum                |
| np.mean() | average            |
| np.max()  | maximum            |
| np.min()  | minimum            |
| np.std()  | standard deviation |

**Example :** 

```python
arr = np.array([10,20,30,40])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
```

## 12. Random Numbers in NumPy

NumPy can generate random numbers.

**Example:**

```python
arr = np.random.rand(3)
print(arr)
```

Random integers:

```python
arr = np.random.randint(1,10,5)
print(arr)
```

## 13. Copy vs View

**Copy**

Creates new array.

```python
a = np.array([1,2,3])
b = a.copy()

b[0] = 100

print(a)
print(b)
```

**View**

Shares memory.

```python
a = np.array([1,2,3])
b = a.view()

b[0] = 100

print(a)
print(b)
```

## 14. Iterating Through Arrays

**Example :**

```python
arr = np.array([1,2,3,4])

for x in arr:
    print(x)
```

## 15. Example (Student Marks Analysis)

**Example :** 

```python
import numpy as np

marks = np.array([70,80,90,60,75])

print("Average:", np.mean(marks))
print("Maximum:", np.max(marks))
print("Minimum:", np.min(marks))
```
