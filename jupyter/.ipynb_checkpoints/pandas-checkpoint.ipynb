# Pandas Tutorial
Pandas is one of the most popular Python libraries for data analysis and data manipulation. It provides easy-to-use data structures like Series and DataFrame to work with structured data such as tables, CSV files, Excel sheets, databases, etc.

Pandas is widely used in data science, machine learning, and analytics.

### 1. Installing Pandas

Install using pip

`pip install pandas`

Import pandas in Python

`import pandas as pd`

## 2. What is a Series?

A Series is a one-dimensional labeled array capable of holding any data type.

Example:

````python
import pandas as pd
data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)
````

**Custom Index**

````python
s = pd.Series([100,200,300], index=["A","B","C"])
print(s)
````

**Access element**

````python
print(s["B"])
````

## 3. What is a DataFrame?

A DataFrame is a two-dimensional table-like structure with rows and columns.

Example

````python
import pandas as pd
data = {
    "Name": ["Ram","Shyam","Hari"],
    "Age": [20,21,22],
    "City": ["Delhi","Mumbai","Pune"]
}
df = pd.DataFrame(data)
print(df)
````

## 4. Reading Data from Files
**Read CSV**

````python
df = pd.read_csv("data.csv")
print(df)
````

**Read Excel**

````python
df1 = pd.read_excel("data.xlsx")  #install openpyxl library (`pip install openpyxl`)
print(df1)
````

**Read JSON**

````python
df = pd.read_json("data.json")
print(df)
````

## 7. Adding New Column

````python
df["Salary"] = 20000
df.insert(2,"HRA",5000) #At specific position
df.head()
````

## Simple Example (Real Dataset)

````python
import pandas as pd

data = {
    "Name":["A","B","C","D"],
    "Marks":[80,90,75,88],
    "Subject":["Math","Math","Science","Science"]
}

df = pd.DataFrame(data)

print(df)

print("Average Marks:", df["Marks"].mean())

print(df.groupby("Subject")["Marks"].mean())
````

**Question:

````python
# Question 1

import pandas as pd

data = {
    "Name": ["Ram", "Shyam", "Hari"],
    "Marks": [80, 90, 85]
}

df = pd.DataFrame(data)

# Insert column Age at position 1
df.insert(1, "Age", [20, 21, 22])

print(df)
````

````python
# Question 2

import pandas as pd

data = {
    "Name": ["Ravi", "Amit", "Sita"],
    "Age": [21, 22, 23]
}

df = pd.DataFrame(data)

df.insert(2, "Country", "India")

print(df)
````

````python
# Question 3

import pandas as pd

data = {
    "Name": ["Ram", "Shyam", "Hari"],
    "Marks": [80, 90, 75]
}

df = pd.DataFrame(data)

df.insert(2, "Result", df["Marks"] >= 80)

print(df)
````

````python
# Question 4

import pandas as pd

data = {
    "Employee": ["A", "B", "C"],
    "Department": ["IT", "HR", "Finance"]
}

df = pd.DataFrame(data)

df.insert(2, "Salary", [30000, 35000, 40000])

print(df)
````
